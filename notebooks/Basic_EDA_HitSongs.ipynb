```python
def compare_hit_flop_all(feature):
    ''' Compares hits vs flops for given 
        feature in dataset'''
    
    x = song_hits['song_year']
    y = song_hits[feature]
    x2 = song_flops['song_year']
    y2 = song_flops[feature]

    fig, axs = plt.subplots(1, 2,figsize=(10,6),sharey=True)

    axs[0].plot(x, y,'o')
    m, b = np.polyfit(x, y, 1)
    axs[0].plot(x,m*x+b)
    axs[0].set_title('{} of Hits Over Time'.format(feature.capitalize()))

    axs[1].plot(x2, y2,'o',c='y')
    m, b = np.polyfit(x2, y2, 1)
    axs[1].plot(x2,m*x2+b)
    axs[1].set_title('{} of Flops Over Time'.format(feature.capitalize()))
```

```python
def compare_hit_flop_genre(feature):
    ''' Compares hits vs flops for given 
        feature by genre in dataset'''
    
    x = song_hits['song_year']
    y = song_hits[feature]
    x2 = song_flops['song_year']
    y2 = song_flops[feature]

    fig, axs = plt.subplots(5, 2, figsize=(10,12),sharey=True)

    axs[0, 0].plot(x[song_hits.genre=='r&b'], y[song_hits.genre=='r&b'],'o')
    m, b = np.polyfit(x[song_hits.genre=='r&b'], y[song_hits.genre=='r&b'], 1)
    axs[0, 0].plot(x[song_hits.genre=='r&b'],m*x[song_hits.genre=='r&b']+b)
    axs[0, 0].set_title('R&B {} Over Time (Hits)'.format(feature.capitalize()))

    axs[0, 1].plot(x2[song_flops.genre=='r&b'], y2[song_flops.genre=='r&b'],'o',c='y')
    m, b = np.polyfit(x2[song_flops.genre=='r&b'], y2[song_flops.genre=='r&b'], 1)
    axs[0, 1].plot(x2[song_flops.genre=='r&b'],m*x2[song_flops.genre=='r&b']+b)
    axs[0, 1].set_title('R&B {} Over Time (Flops)'.format(feature.capitalize()))

    axs[1, 0].plot(x[song_hits.genre=='rap'], y[song_hits.genre=='rap'],'o')
    m, b = np.polyfit(x[song_hits.genre=='rap'], y[song_hits.genre=='rap'], 1)
    axs[1, 0].plot(x[song_hits.genre=='rap'],m*x[song_hits.genre=='rap']+b)
    axs[1, 0].set_title('Rap {} Over Time (Hits)'.format(feature.capitalize()))

    axs[1, 1].plot(x2[song_flops.genre=='rap'], y2[song_flops.genre=='rap'],'o',c='y')
    m, b = np.polyfit(x2[song_flops.genre=='rap'], y2[song_flops.genre=='rap'], 1)
    axs[1, 1].plot(x2[song_flops.genre=='rap'],m*x2[song_flops.genre=='rap']+b)
    axs[1, 1].set_title('Rap {} Over Time (Flops)'.format(feature.capitalize()))

    axs[2, 0].plot(x[song_hits.genre=='pop'], y[song_hits.genre=='pop'],'o')
    m, b = np.polyfit(x[song_hits.genre=='pop'], y[song_hits.genre=='pop'], 1)
    axs[2, 0].plot(x[song_hits.genre=='pop'],m*x[song_hits.genre=='pop']+b)
    axs[2, 0].set_title('Pop {} Over Time (Hits)'.format(feature.capitalize()))

    axs[2, 1].plot(x2[song_flops.genre=='pop'], y2[song_flops.genre=='pop'],'o',c='y')
    m, b = np.polyfit(x2[song_flops.genre=='pop'], y2[song_flops.genre=='pop'], 1)
    axs[2, 1].plot(x2[song_flops.genre=='pop'],m*x2[song_flops.genre=='pop']+b)
    axs[2, 1].set_title('Pop {} Over Time (Flops)'.format(feature.capitalize()))

    axs[3, 0].plot(x[song_hits.genre=='latin'], y[song_hits.genre=='latin'],'o')
    m, b = np.polyfit(x[song_hits.genre=='latin'], y[song_hits.genre=='latin'], 1)
    axs[3, 0].plot(x[song_hits.genre=='latin'],m*x[song_hits.genre=='latin']+b)
    axs[3, 0].set_title('Latin {} Over Time (Hits)'.format(feature.capitalize()))

    axs[3, 1].plot(x2[song_flops.genre=='latin'], y2[song_flops.genre=='latin'],'o',c='y')
    m, b = np.polyfit(x2[song_flops.genre=='latin'], y2[song_flops.genre=='latin'], 1)
    axs[3, 1].plot(x2[song_flops.genre=='latin'],m*x2[song_flops.genre=='latin']+b)
    axs[3, 1].set_title('Latin {} Over Time (Flops)'.format(feature.capitalize()))

    axs[4, 0].plot(x[song_hits.genre=='rock'], y[song_hits.genre=='rock'],'o')
    m, b = np.polyfit(x[song_hits.genre=='rock'], y[song_hits.genre=='rock'], 1)
    axs[4, 0].plot(x[song_hits.genre=='rock'],m*x[song_hits.genre=='rock']+b)
    axs[4, 0].set_title('Rock {} Over Time (Hits)'.format(feature.capitalize()))

    axs[4, 1].plot(x2[song_flops.genre=='rock'], y2[song_flops.genre=='rock'],'o',c='y')
    m, b = np.polyfit(x2[song_flops.genre=='rock'], y2[song_flops.genre=='rock'], 1)
    axs[4, 1].plot(x2[song_flops.genre=='rock'],m*x2[song_flops.genre=='rock']+b)
    axs[4, 1].set_title('Rock {} Over Time (Flops)'.format(feature.capitalize()))

    fig.tight_layout()
```
